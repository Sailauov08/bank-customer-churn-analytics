```python

import sqlite3
import pandas as pd
from sklearn.datasets import fetch_openml

print("Халықаралық серверден елдер мен жынысы бар деректерді жүктеу...")
bank_data = fetch_openml(name='churn-for-bank-customers', version=1, as_frame=True, parser='auto')
df = bank_data.frame

# Елдер (Geography) мен Жынысын (Gender) бағандар тізіміне қосамыз
df = df[['CustomerId', 'Surname', 'CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']]
df.columns = ['customer_id', 'surname', 'credit_score', 'geography', 'gender', 'age', 'tenure', 'balance', 'num_products', 'has_credit_card', 'is_active_member', 'estimated_salary', 'churn']

conn = sqlite3.connect('bank_analytics.db')
df.to_sql('raw_customers', conn, if_exists='replace', index=False)
conn.close()

df.to_excel('raw_bank_data.xlsx', index=False)
print(f"Дайын! Жаңаланған {df.shape[0]} жол базаға жазылды.")
```
