```python
import sqlite3
import pandas as pd
from sklearn.datasets import fetch_openml

print("Халықаралық серверден деректерді жүктеу басталуда...")
bank_data = fetch_openml(name='churn-for-bank-customers', version=1, as_frame=True, parser='auto')
df = bank_data.frame

# Бағандарды реттеу
df = df[['CustomerId', 'Surname', 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']]
df.columns = ['customer_id', 'surname', 'credit_score', 'age', 'tenure', 'balance', 'num_products', 'has_credit_card', 'is_active_member', 'estimated_salary', 'churn']

# 1. SQL базасына ('raw_customers' кестесіне) жіберу
conn = sqlite3.connect('bank_analytics.db')
df.to_sql('raw_customers', conn, if_exists='replace', index=False)
conn.close()

# 2. Шикі деректі Excel файлына жіберу
df.to_excel('raw_bank_data.xlsx', index=False)

# 3. Нақты есеп беру
print("\n" + "="*50)
print(f"📊 1-ҚАДАМ ЕСЕБІ: Серверден барлығы {df.shape[0]} клиенттің дерегі ЖИНАЛДЫ.")
print("💾 Деректер 'raw_customers' SQL кестесіне ЖӘНЕ 'raw_bank_data.xlsx' файлына сәтті жіберілді!")
print("="*50)
```
