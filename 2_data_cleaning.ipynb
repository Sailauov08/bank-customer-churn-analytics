```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('bank_analytics.db')
df_raw = pd.read_sql_query("SELECT * FROM raw_customers", conn)

df_cleaned = df_raw.copy()
df_cleaned['surname'] = df_cleaned['surname'].astype(str)
df_cleaned['geography'] = df_cleaned['geography'].astype(str)
df_cleaned['gender'] = df_cleaned['gender'].astype(str)
df_cleaned = df_cleaned.dropna()

df_cleaned.to_sql('cleaned_customers', conn, if_exists='replace', index=False)
conn.close()

# Осы Excel файлы біздің Power BI-ға негіз болады!
df_cleaned.to_excel('cleaned_bank_data.xlsx', index=False)

print(f"Тазалау аяқталды! 'cleaned_bank_data.xlsx' файлы Power BI-ға дайын.")
```
