```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('bank_analytics.db')
df_raw = pd.read_sql_query("SELECT * FROM raw_customers", conn)

# Тазалау процесі
df_cleaned = df_raw.copy()
df_cleaned['surname'] = df_cleaned['surname'].astype(str)
df_cleaned = df_cleaned.dropna() # Бос жолдар болса өшіру

# 1. Тазаланған деректі жаңа SQL кестесіне ('cleaned_customers') жіберу
df_cleaned.to_sql('cleaned_customers', conn, if_exists='replace', index=False)
conn.close()

# 2. Тазаланған деректі жаңа Excel файлына жіберу
df_cleaned.to_excel('cleaned_bank_data.xlsx', index=False)

# 3. Салыстырмалы есеп беру
print("\n" + "="*50)
print(f"🧹 2-ҚАДАМ ЕСЕБІ: Тазалауға дейін {df_raw.shape[0]} жол бар еді.")
print(f"✅ Тазалаудан кейін {df_cleaned.shape[0]} жол ҚАЛДЫ.")
print(f"❌ Өшірілген (бос немесе қате) жолдар саны: {df_raw.shape[0] - df_cleaned.shape[0]}")
print("💾 Тазаланған база 'cleaned_customers' SQL кестесіне ЖӘНЕ 'cleaned_bank_data.xlsx' файлына жіберілді!")
print("="*50)
```
